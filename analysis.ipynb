```python
from data.flights import flights
from bokeh.charts import output_notebook, show
from bokeh.charts import Histogram, BoxPlot
from bokeh.models.widgets.tables import DataTable, TableColumn, NumberFormatter
from bokeh.models import ColumnDataSource
```

# Which Paper Airplane Flies the Furthest?
By Team Caraway

```python
output_notebook(hide_banner=True)
```

```python
boxes = BoxPlot(flights, values='distance (ft)', color='type', label='type')
show(boxes)
```

## Model: Dart
![image](https://www.evernote.com/l/ANq56FIF5O9CYbZRdFBadZm94AK5sbJjFFgB/image.jpg)

### Flight Data

```python
dart_flights = flights[flights['type'] == 'dart']
columns = [
    TableColumn(field='id', title='ID'),
    TableColumn(field='type', title='Model'),
    TableColumn(field='distance (ft)', title='Distance (feet)', formatter=NumberFormatter(format='0,0.00', rounding='round'))
]

dart_table = DataTable(source=ColumnDataSource(dart_flights), columns=columns)
show(dart_table)
```

```python
dart_plot = Histogram(dart_flights, title="Dart Flight distance", values="distance (ft)", bins=list(range(10, 32, 2)))
show(dart_plot)
```
